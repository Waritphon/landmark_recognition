```python
import numpy as np
import pandas as pd
```

```python
train_selected_label_to_category = pd.read_csv('train/train_selected_label_to_category.csv')
```

```python
recognition_solution = pd.read_csv("test/recognition_solution_v2.1.csv")
recognition_solution = recognition_solution.drop(columns="Usage")
```

```python
recognition_solution
```

```python
recognition_solution = recognition_solution[recognition_solution["landmarks"].notnull()]
```

```python
recognition_solution
```

```python
recognition_solution = recognition_solution[~recognition_solution["landmarks"].str.contains(' ')]
recognition_solution["landmarks"] = recognition_solution["landmarks"].astype(np.int64)
```

```python
recognition_solution
```

```python
recognition_solution = recognition_solution.merge(train_selected_label_to_category[["landmark_id", "label"]], \
                                                  left_on="landmarks", right_on="landmark_id", how="right")
```

```python
recognition_solution
```

```python
recognition_solution = recognition_solution.drop(columns=["landmarks", "landmark_id"]).dropna().reset_index()
```
