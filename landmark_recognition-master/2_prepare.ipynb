```python
train = pd.read_csv("data/data.csv")
train_label_to_category = pd.read_csv("data/label_to_category.csv")
```

# Data Split

```python
#ทำการแยก data โดยแยกตาม Label อ่านจาก csv นะจะเป็น 
# 1. label มี 100 อันและแต่ละอันบอกว่าตัวเองทำอะไรอยู่ 100 Class 
# 2. label จะใช้ในการจัดการกับข้อมูลในนี้ต่อไป

train = train.groupby('label').filter(lambda x : len(x) >= 500)
train = train.merge(train_label_to_category, on='label', how='inner')
train['category'] = train['category'].str.replace('https://upload.wikimedia.org/wikipedia/commons/Category:', '').str.lower()
```

```python
train.head()
```

# เเสดงข้อมูล 100 Class ที่สนใจ

```python
train['category'].value_counts()
```

```python
keywords = 'mount|lake|fall|cave|canyon|park|garden|cemetery|castle|palace|monastery|museum|fort|port|tower|library|monument|shrine|church|bridge'
train_selected = train[train['category'].str.contains(keywords, regex=True) & ~train['category'].str.contains('%')]
train_selected = train_selected.reset_index(drop=True)
train_selected['category'].value_counts()
```

```python
train_selected.head()
```

```python
train_selected_label_to_category = train_selected[['landmark_id_y', 'category']].drop_duplicates()
train_selected_label_to_category = train_selected_label_to_category.reset_index(drop=True)
train_selected_label_to_category['label'] = train_selected_label_to_category.index
train_selected_label_to_category.head()
```

```python
train_selected = train_selected.merge(train_selected_label_to_category[['landmark_id_y', 'label']], on='landmark_id_y', how='left')
train_selected = train_selected.reset_index()
train_selected.head()
```

```python
train_selected = train_selected.drop(columns=['id', 'landmark_id', 'category'])
train_selected.head()
```

```python
train_selected.to_csv('data/data.csv', index=False)
train_selected_label_to_category.to_csv('data/label_to_category.csv', index=False)
```
